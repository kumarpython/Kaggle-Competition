```python
x = data.iloc[:,1:-1]
y = data.iloc[:,-1]
m = len(y) # no. of training samples
x
```

```python
#adding ones to X
one = np.ones((m,1))
X = np.append(one, x, axis=1)

#reshape Y to a column vector
y = np.array(y).reshape((m,1))
X
```

**Since it is a comparatibly small dataset, we can use Normal Eqn...
it has no categorical features, so FE not needed**

```python
# The NORMAL Equation
def normal_equation(X, Y):
  beta = np.dot((np.linalg.inv(np.dot(X.T,X))), np.dot(X.T,Y))
  return beta

normal_equation(X,y)
```
